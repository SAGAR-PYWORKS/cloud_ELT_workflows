###Modifying Column Names

```python
import re
from pyspark.sql import DataFrame

def clean_column_names(df: DataFrame) -> DataFrame:
    # Create a mapping of old to new column names
    new_columns = []
    for col_name in df.columns:
        clean_name = col_name.replace('\n', ' ')           # remove newlines
        clean_name = re.sub(r'[^\w\s]', '', clean_name)    # remove special chars
        clean_name = clean_name.strip().lower()            # lowercase and strip
        clean_name = re.sub(r'\s+', '_', clean_name)       # spaces to underscores
        new_columns.append(clean_name)
    
    # Rename columns in the DataFrame
    for old_name, new_name in zip(df.columns, new_columns):
        df = df.withColumnRenamed(old_name, new_name)
    
    return df
```

##1. Clean Numerical Columns

```python
from pyspark.sql.functions import regexp_replace, col
from pyspark.sql.types import DoubleType

def cleanse_numerical_columns(df: DataFrame, columns: list) -> DataFrame:
    for col_name in columns:
        # Keep only digits, minus, and decimal point
        df = df.withColumn(
            col_name,
            regexp_replace(col(col_name), "[^0-9.-]", "").cast(DoubleType())
        )
    return df
```

##2. Changed Date Format from mm/dd/yyyy to yyyy-mm-dd

```python
from pyspark.sql.functions import col, regexp_replace, when, length, split, concat_ws, lpad, to_date, date_format,lit

def fix_two_digit_years(df, date_columns):
    for col_name in date_columns:
        # Replace '/' with '-' to simplify splitting
        df = df.withColumn(col_name, regexp_replace(col(col_name).cast("string"), "/", "-"))
        
        # Split date into parts
        parts = split(col(col_name), "-")

        # Assume last part is year. If it's 2 digits, prepend '20'
        fixed_year = when(length(parts.getItem(2)) == 2, concat_ws("", lit("20"), lpad(parts.getItem(2), 2, "0")))\
                     .otherwise(parts.getItem(2))  # 12/23/24 Output: 2024

        # Reconstruct the fixed date string
        fixed_date_str = concat_ws("-", parts.getItem(0), parts.getItem(1), fixed_year) #Output:2024/12/23

        # Convert to actual date
        df = df.withColumn(col_name, date_format(to_date(fixed_date_str, "MM-dd-yyyy"), "yyyy-MM-dd"))
    
    return df
```

```python
spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")

date_columns = [
    'date_opened', 'date_closed', 'date_reopened', 'date_reclosed',
    'cert_start', 'cert_end', 'loss_date', 'date_reported',
    'date_claim_acknowledged', 'date_reserve_first_established',
    'peer_review_date', 'diary_date', 'date_ind_paid', 'date_exp_paid',
    'salvage_date', 'subro_date'
]

df1_date_cleaned2 = fix_two_digit_years(df1_numeric_cleaned, date_columns)
df1_date_cleaned2.select("date_opened", "date_closed", "loss_date").show(truncate=False)
```

##3. Removed leading and trailing spaces and changed to upperCase

```python
from pyspark.sql.functions import trim, upper, col

def cleanse_string_columns(df):
    string_cols = [
        'insured_name',
        'insured_address',
        'driver_name',
        'loss_location',
        'cause_of_loss',
        'nature_of_loss',
        'vehicle_1_description',
        'vehicle_2_description'
    ]
    
    for col_name in string_cols:
        df = df.withColumn(col_name, upper(trim(col(col_name))))
    
    return df
```

```python
df_cleaned_string = cleanse_string_columns(df1_date_cleaned2)
display(df_cleaned_string)
```
